```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("C:/Users/HP/OneDrive/Desktop/cirrhosis.csv")
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data['Age'], bins=20, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Drug', y='Bilirubin', data=data)
plt.title('Bilirubin levels by Drug')
plt.xlabel('Drug')
plt.ylabel('Bilirubin')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.countplot(x='Status', data=data)
plt.title('Count of Status')
plt.xlabel('Status')
plt.ylabel('Count')
plt.show()
```

```python
correlation_matrix = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Ascites', y='Albumin', data=data)
plt.title('Albumin levels by Ascites')
plt.xlabel('Ascites')
plt.ylabel('Albumin')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Bilirubin', y='Cholesterol', data=data)
plt.title('Bilirubin vs. Cholesterol')
plt.xlabel('Bilirubin')
plt.ylabel('Cholesterol')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Drug', hue='Sex', data=data)
plt.title('Drug distribution by Gender')
plt.xlabel('Drug')
plt.ylabel('Count')
plt.show()
```

```python
sns.pairplot(data, vars=['Bilirubin', 'Cholesterol', 'Albumin', 'Copper'], hue='Status')
plt.title('Pairplot of Numeric Variables')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Stage', data=data)
plt.title('Stage distribution')
plt.xlabel('Stage')
plt.ylabel('Count')
plt.show()
```

```python
pd.crosstab(data['Drug'], data['Status']).plot(kind='bar', stacked=True)
plt.title('Drug and Status')
plt.xlabel('Drug')
plt.ylabel('Count')
plt.show()
```
